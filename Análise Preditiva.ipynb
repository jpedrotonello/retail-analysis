```python
import pandas as pd
from pycaret.regression import *
import seaborn as sns
```

```python
df = pd.read_csv('WALMART_SALES_DATA.csv')
print(df.dtypes)
df.head()
```

```python
# obtendo métricas de desempenho da regressão linear
s = setup(df[['Store', 'Weekly_Sales']], target='Weekly_Sales',
          categorical_features=['Store'], session_id=0)
lr_model = create_model('lr')
```

```python
# obtendo métricas de desempenho de vários modelos
s = setup(df[['Store', 'Weekly_Sales']], target='Weekly_Sales',
          categorical_features=['Store'], session_id=0)
model_only_store = compare_models()
```

```python
# Testando modelos com todas as variáveis analisadas
s = setup(df, target = 'Weekly_Sales', ignore_features=['Date'], session_id=0)
best_model = compare_models(['lightgbm', 'rf', 'dt'])
```

```python
# Avaliação dos modelos
print(best_model)
evaluate_model(best_model)
```
